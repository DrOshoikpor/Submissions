## What was the moving average of the various stocks?



* A moving average (MA) is a widely used technical indicator that smooths out price trends by filtering out the “noise” from random short-term price fluctuations. We will need the MA in this project to identify trend direction and to determine support and resistance levels. Since,a simple moving average (SMA) is a rolling mean of the recent prices over a specific number of time periods (e.g. 10 days, 20 days, 50 days etc), we will come up with a list named " moving_average_day" that will take in 10, 20, 50 days respectively.

Please read more here:


[Moving Averages](https://www.investopedia.com/articles/active-trading/052014/how-use-moving-average-buy-stocks.asp)

## Value at Risk Using the Monte Carlo Method

Under the Monte Carlo method, Value at Risk is calculated by randomly creating a number of scenarios for future rates using non-linear pricing models to estimate the change in value for each scenario, and then calculating the VaR according to the worst losses.

The Monte Carlo method is suitable for a great range of risk measurement problems, especially when dealing with complicated factors. It assumes that there is a known probability distribution for risk factors.

For this project, here is what is relevant to us:


firstly, use the Geometric Brownian Motion (GBM), which is technically known as a Markov process. This means that the stock price follows a random walk and is consistent with (at the very least) the weak form of the Efficient Market Hypothesis (EMH): past price information is already incorporated and the next price movement is "conditionally independent" of past price movements.

This means that the past information on the price of a stock is independent of where the stock price will be in the future, basically meaning, you can't perfectly predict the future solely based on the previous price of a stock.

Now we see that the change in the stock price is the current stock price multiplied by two terms. The first term is known as "drift", which is the average daily return multiplied by the change of time. The second term is known as "shock", for each time period the stock will "drift" and then experience a "shock" which will randomly push the stock price up or down. By simulating this series of steps of drift and shock thousands of times, we can begin to do a simulation of where we might expect the stock price to be.

Please read more about this method here:  
[Monte Carlo Method Explanation](https://www.investopedia.com/articles/07/montecarlo.asp)

[More References on Value At Risk](https://corporatefinanceinstitute.com/resources/knowledge/trading-investing/value-at-risk-var/)

[More Recources on Monte Carlo](https://www.interviewqs.com/ddi-code-snippets/intro-monte-carlo)

```python
## This is an attempt to first define all the variables to be used for th Amazon stock

# time gorizon set at;
days = 365

# take delta to;
dt = 1/days

# Now let's take our mu (drift) from the expected return data we got for AMAZON
mu = rets.mean()['AMZN']

# Now let's quickly take the volatility of the stock from the std() of the average return for AMAZON
sigma = rets.std()['AMZN']

# By the way, Stock market volatility is a measure of how much the stock market's overall value fluctuates up and down.

#Adapted from https://www.fool.com/investing/how-to-invest/stocks/stock-market-volatility/
```

```python
def stock_monte_carlo(start_price,days,mu,sigma): ### Remeber mu = drift
    ''' This function takes in start stock price, days of simulation,mu,sigma, and returns simulated price array'''
    
    # Define a price array
    price = np.zeros(days)
    price[0] = start_price
    
    # Shock and Drift
    shock = np.zeros(days)
    drift = np.zeros(days)
    
    # Run price array for number of days
    for x in range(1,days):
        
        # Calculate Schock
        shock[x] = np.random.normal(loc=mu * dt, scale=sigma * np.sqrt(dt))
        # Calculate Drift
        drift[x] = mu * dt
        # Calculate Price
        price[x] = price[x-1] + (price[x-1] * (drift[x] + shock[x]))
        
    return price
```

```python
# Starting with AMAZON Stock
start_price = 745.510010

# Set a large numebr of runs
runs = 10000

# Create an empty matrix to hold the end price data
simulations = np.zeros(runs)

for run in range(runs):    
    # Set the simulation data point as the last stock price for that run
    simulations[run] = stock_monte_carlo(start_price,days,mu,sigma)[days-1]
```

An array of simulation is established, now lets plot a histogram and use quartile to define the risk for the stocks

```python
# Now we will define q as the 1% empirical quantile, this basically means that 99% of the values should fall between q = np.percentile(simulations,1)

q = np.percentile(simulations,1)

# Now let's plot the distribution of the end prices
plt.hist(simulations, bins=200)

# Using plt.figtext to fill in some additional information onto the plot

# starting price
plt.figtext(0.6,0.8, s='Start Price: $%.2f' % start_price)

# mean ending price
plt.figtext(0.6,0.7, s='Mean Final Price: $%.2f' % simulations.mean())

# Variance of the price (within 99% confidence interval)
plt.figtext(0.6,0.6, s='VaR(0.99): $%.2f' % (start_price - q))

# To display 1% quantile
plt.figtext(0.15, 0.6, s="q(0.99): $%.2f" % q)

# Plot a line at the 1% quantile result
plt.axvline(x=q, linewidth=4, color='r')

# For plot title
plt.title(label="Final price distribution for AMAZON Stock(AMZN) after %s days" % days, weight='bold')
```

This basically means for every initial AMAZON stock one purchase, they are putting about $29.49 at risk 99% of the time from our Monte Carlo Simulation.

```python
# For SQUARE tock Price
start_price = 12.12

# Set a large numebr of runs
runs = 10000

# Create an empty matrix to hold the end price data
simulations = np.zeros(runs)

for run in range(runs):    
    # Set the simulation data point as the last stock price for that run
    simulations[run] = stock_monte_carlo(start_price,days,mu,sigma)[days-1]
```

```python
# define q as the 1% empirical quantile, this basically means that 99% of the values should fall between q = np.percentile(simulations,1)
q = np.percentile(simulations,1)

# Now let's plot the distribution of the end prices
plt.hist(simulations, bins=200)

# Using plt.figtext to fill in some additional information onto the plot

# starting price
plt.figtext(0.6,0.8, s='Start Price: $%.2f' % start_price)

# mean ending price
plt.figtext(0.6,0.7, s='Mean Final Price: $%.2f' % simulations.mean())

# Variance of the price (within 99% confidence interval)
plt.figtext(0.6,0.6, s='VaR(0.99): $%.2f' % (start_price - q))

# To display 1% quantile
plt.figtext(0.15, 0.6, s="q(0.99): $%.2f" % q)

# Plot a line at the 1% quantile result
plt.axvline(x=q, linewidth=4, color='r')

# For plot title
plt.title(label="Final price distribution for SQUARE Stock(SQ) after %s days" % days, weight='bold')
```

**interpretation**


* The visual have shown 1% empirical quantile of the final price distribution to estimate the Value at Risk for the SQUARE Stock(SQ), which looks to be $0.48 for every investment of 12.12 (The price of one initial Amazon Stock).

This basically means for every initial SQUARE stock one purchase, they are putting about $0.48 at risk 99% of the time from our Monte Carlo Simulation.

## Predicting stock prices by using Linear Regression model



We used the Linear regression(LR) model to try to predict the closing stock price of all the stocks used for this project.Linear Regression is the supervised Machine Learning model in which the model finds the best fit linear line between the independent and dependent variable i.e it finds the linear relationship between the dependent and independent variable.

   - First we defined features for Independent variables for all stocks and named it (X) and y for dependent variables.
    
   - Split the data into 80% training and 20% testing and instantiate the linear regression model for each stock.
    
   - We fit the model for all stocks, generated values for coefficient and intercept.
    
   - Printed predicted values and placed it in a dataframe with actual price values for ease of comparison.
    
   - We determined how our model performed by evaluating with metrices such as Mean Absolute Error,Root Squared Mean Error,R-Squared,Mean Squared Error and cross Val, Score.
    
   - Define a variable named graph for all stocks and visualization.
    
   - Plotted a graph to compare predicted price and actual closing price for all stocks.

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn import metrics
```

```python
AMZN
```

```python
AMZN.describe()
```

```python
X = AMZN[['High', 'Low', 'Open', 'Volume']].values
y = AMZN['Close'].values
```

```python
# Split the data into 80% training and 20% testing
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# instantiate
lr_AMZN = LinearRegression()
```

```python
# fit
lr_AMZN.fit(X_train, y_train)
```

```python
lr_AMZN.coef_
```

```python
print(lr_AMZN.intercept_)
```

```python
predicted_AMZN = lr_AMZN.predict(X_test)
```

```python
print(predicted_AMZN)
```

```python
df_AMZN = pd.DataFrame({'Actual' :y_test.flatten(), 'predicted' :predicted_AMZN.flatten()})
```

```python
df_AMZN.head(30)
```

```python
# how correct are the predictions? Evaluation metrics
import math
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, predicted_AMZN))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, predicted_AMZN))
print('Root Mean Squared Error:',math.sqrt(metrics.mean_squared_error(y_test, predicted_AMZN)))
```

**Comment**
 - The Mean Absolute error(MAE) tells us how big of an error we can expect from the forecast on average. from our result, we can expect 9.8 error
 - The value of our root mean square error(RMSE) is 14.8 standard deviation of the unexplained variance. If we compare this to the mean value of all the closing price that is 2026.862183, the gap is very huge hence our algorithm  can still make reasonably good predictions.
 - Mean Squared Error has 219.5 value that tells us how close we are to finding the line of best fit. usually the smaller the better and our model did pretty good.

```python
#R^2 from model
lr_AMZN.score(X_test, y_test)
```

**Interpretation**

The R-Squared shows if the model is a good fit predicted values and how good of a “fit” it is.The higher the better and our algorithm shows 0.9996987573528978.This means that 99.97% of the data fits the regression line which is a fantastic fit for the model.

```python
cross_val_score(lr_AMZN, X_test, y_test, cv=5).mean()
```

**Interpretation**

The cross validation shows that our model can evaluate the performance of unseen data at 99.97% validation score. This usually assist with preventing overfitting.
